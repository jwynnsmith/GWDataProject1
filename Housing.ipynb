```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Store filepath in a variable
housing_file = "DC_HousingN1_39.csv"
cluster_file = "clusters_names.csv"
dc_crime_file = "my_crime_data.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
# This is now a data frame
housing_df = pd.read_csv(housing_file, encoding="ISO-8859-1")
housing_df = housing_df.loc[housing_df["Year"] > 2009, ["Cluster", "Year", "Sales of Single-Family Homes", "Price of Single-Family Homes (2016 $)", "Food stamp recipients"]]
housing_df.head()
```

```python
name_df = pd.read_csv(cluster_file, encoding="ISO-8859-1")
name_df.head()
```

```python
cluster_list = housing_df["Cluster"].unique().tolist()
cluster_list
```

```python
name_df.style.set_properties(**{'text-align': 'left'})

name_df["Clusters"] = name_df["Clusters"].str[2:]

name_df["Cluster_ID"] = cluster_list

name_df
```

```python
name_df = name_df.set_index("Cluster_ID")
name_df
```

```python
housing_df["Single-Family Homes Median Price (2016 $)"] = housing_df["Price of Single-Family Homes (2016 $)"].str.replace(',', '').dropna().astype("int")
housing_df.head()
```

```python
housing_df["Food Stamp Count"] = housing_df["Food stamp recipients"].dropna().str.replace(',', '').dropna().astype("int")
housing_df.head()
```

```python
neighborhood_stats_df = housing_df[["Cluster", "Year", "Sales of Single-Family Homes", "Single-Family Homes Median Price (2016 $)", "Food Stamp Count"]]
neighborhood_stats_df
```

```python
crime_df = pd.read_csv(dc_crime_file, encoding="ISO-8859-1")

crime_df["END_DATE"] = crime_df["END_DATE"].str[:4]
crime_df["START_DATE"] = crime_df["START_DATE"].str[:4]
crime_df.head()
```

```python
crime_subset_df = crime_df[["NEIGHBORHOOD_CLUSTER","OFFENSE","START_DATE"]]
crime_subset_df["START_DATE"] = crime_subset_df["START_DATE"].str[:4]


crime_subset_df.head()
```

```python
dc_crime_df = crime_subset_df.rename(columns={"NEIGHBORHOOD_CLUSTER":"Cluster","OFFENSE":"Crime", "START_DATE":"Year"})
dc_crime_df["Cluster"] = dc_crime_df["Cluster"].str.replace("cluster "," ").dropna().astype(int)
dc_crime_df["Year"] = dc_crime_df["Year"].dropna().astype(float)
dc_crime_df["Year"] = dc_crime_df["Year"].apply(np.floor)
dc_crime_df = dc_crime_df[["Cluster","Year","Crime"]]
dc_crime_df = dc_crime_df.loc[dc_crime_df["Year"] > 2009, ["Cluster", "Year", "Crime"]]
dc_crime_df = dc_crime_df.loc[dc_crime_df["Year"] < 2017, ["Cluster", "Year", "Crime"]]
dc_crime_df
```
